## Q1.a

**(a) $\sum_{i=0}^{n}i^{2}=\frac{n(n+1)(2n+1)}{6}$**


**1. Caso Base ($n = n_0 = 1$):**
  
* Lado Esquerdo : $\sum_{i=0}^{1}i^{2} = 0^2 + 1^2 = 1$


* Lado Direito : $\frac{1(1+1)(2(1)+1)}{6} = \frac{1(2)(3)}{6} = 1$


* Lado Esquerdo = Lado Direito


* Como obtemos (Lado Esquerdo = Lado Direito), o caso base é válido.




**2. Hipótese de indução:**

* Assume-se que P(k) é verdadeira para um $k \ge n0$:

    $\sum_{i=0}^{k}i^{2}=\frac{k(k+1)(2k+1)}{6}$



**3. Passo Indutivo:**

Provar que a fórmula é válida para k+1. Substituindo por k+1 temos:

$\sum_{i=0}^{k+1}i^{2}=\frac{(k+1)((k+1)+1)(2(k+1)+1)}{6}$


Começando com o lado esquerdo temos:

$\sum_{i=0}^{k+1}i^{2} = \sum_{i=0}^{k}i^{2} + (k+1)^2 = \frac{k(k+1)(2k+1)}{6} + (k+1)^2 = \frac{k(k+1)(2k+1) + 6(k+1)^2}{6} $


Fatorando (k+1):


$ \frac{k(k+1)(2k+1) + 6(k+1)^2}{6} = \frac{(k+1)[k(2k+1) + 6(k+1)]}{6} = \frac{(k+1)[2k^2 + k + 6k + 6]}{6} $



Simplificando temos:


$\frac{(k+1)[2k^2 + k + 6k + 6]}{6}  = \frac{(k+1)(2k^2 + 7k + 6)}{6} = \frac{(k+1)(k+2)(2k+3)}{6}$



Reescrevendo temos:

$\frac{(k+1)(k+2)(2k+3)}{6} = \frac{(k+1)((k+1)+1)(2(k+1)+1)}{6}$



-> chegamos na mesma equação do lado direito.

Concluindo, pelo princípio da indução matemática, a fórmula $\sum_{i=0}^{n}i^{2}=\frac{n(n+1)(2n+1)}{6}$ é válida para todo $n \ge 1$.



## Q1.b
(b) $\sum_{i=0}^{n}i^{3}=\frac{n^{2}(n+1)^{2}}{4}$ para $n \ge 1$



**1. Caso Base ($n = n_0 = 1$):**

* Lado Esquerdo: $\sum_{i=0}^{1}i^{3} = 0^3 + 1^3 = 1$
* Lado Direito: $\frac{1^2(1+1)^2}{4} = \frac{1(2^2)}{4} = \frac{4}{4} = 1$
* Lado Esquerdo = Lado Direito, o caso base é válido.




**2. Hipótese Indutiva:**

* Assume-se que P(k) é verdadeira para um $k \ge n0$:

$\sum_{i=0}^{k}i^{3}=\frac{k^{2}(k+1)^{2}}{4}$



**3. Passo Indutivo:**


Provar que a fórmula é válida para k+1. Substituindo por k+1 temos:

$\sum_{i=0}^{k+1}i^{3}=\frac{(k+1)^{2}((k+1)+1)^{2}}{4}$


Começando com o lado esquerdo temos:

$\sum_{i=0}^{k+1}i^{3} = \sum_{i=0}^{k}i^{3} + (k+1)^3 = \frac{k^{2}(k+1)^{2}}{4} + (k+1)^3 = \frac{k^{2}(k+1)^{2} + 4(k+1)^3}{4}$


Fatorando $(k+1)^2$:

$ \frac{k^{2}(k+1)^{2} + 4(k+1)^3}{4} = \frac{(k+1)^{2}[k^2 + 4(k+1)]}{4} = \frac{(k+1)^{2}[k^2 + 4k + 4]}{4}$


Reescrevendo temos:

$ \frac{(k+1)^{2}[k^2 + 4k + 4]}{4} = \frac{(k+1)^{2}(k+2)^{2}}{4} = \frac{(k+1)^{2}((k+1)+1)^{2}}{4} $


-> chegamos na mesma equação do lado direito.

Concluindo, pelo princípio da indução matemática, a fórmula  $\sum_{i=0}^{n}i^{3}=\frac{n^{2}(n+1)^{2}}{4}$ é válida para todo $n \ge 1$.




## Q2.a

Para a recorrência

$
T(1)=k,
\quad
T(n)=c\,T(n-1),
$

onde

* **n** é o tamanho do problema,
* **T(n)** é o custo total,
* **c** e **k** são constantes.



**1. Caso base**

Para $n=1$:

$
T(1)=k.
$



**2. Passo recursivo**

Para $n>1$:

$
T(n)=c\,T(n-1).
$



**3. Método da iteração**

Começando com $T(1)$ temos:

* $T(2)=c\,T(1)=c\cdot k$
* $T(3)=c\,T(2)=c\,(c\cdot k)=c^2\,k$
* $T(4)=c\,T(3)=c\,(c^2\,k)=c^3\,k$
* …
* Após $n-1$ iterações:

  $
    T(n)=k\,c^{\,n-1}.
  $



**4. Exemplo**

Se $c=2$ e $k=3$:

* $T(1)=3$
* $T(2)=2\cdot3=6$
* $T(3)=2\cdot6=12$
* $T(4)=2\cdot12=24$

Logo, $T(n)=3\cdot2^{\,n-1}.$



**Conclusão:**

A solução fechada é

$
\boxed{T(n)=k\,c^{\,n-1},}
$

isto é, $T(n)=\Theta(c^n)$.


## Q2.b

\quad

$

onde **n** é o tamanho do problema, **T(n)** o custo total, há 3 chamadas de tamanho $n/4$ e custo de combinação $n^2$.

**1. Caso base**

Para $n=1$:

$
T(1)=1.
$

Queremos garantir $T(1)\le A\cdot1^2$. Basta escolher $A\ge1$.

**2. Hipótese de indução**

Suponhamos que, para todo $k<n$,

$
T(k)\le A\,k^{2}.
$

**3. Passo indutivo**

Para $n>1$:

1. Começando com a definição:

   $
     T(n)=3\,T(n/4)+n^{2}.
   $

2. Usando a hipótese em $T(n/4)$:

   $
     T(n)\le 3\bigl(A\,(n/4)^{2}\bigr)+n^{2}.
   $

3. Desenvolvendo:

   $
     =3A\frac{n^{2}}{16}+n^{2}
     =\frac{3A}{16}\,n^{2}+n^{2}.
   $

4. Junta termos semelhantes:

   $
     =\Bigl(\frac{3A}{16}+1\Bigr)\,n^{2}.
   $

5. Para ter $\bigl(\tfrac{3A}{16}+1\bigr)\le A$:

   $
     \frac{3A}{16}+1\le A
     \quad\Longrightarrow\quad
     1\le A-\frac{3A}{16}
     =\frac{13A}{16}
     \quad\Longrightarrow\quad
     A\ge\frac{16}{13}.
   $

6. Escolhendo $A=2$ (que satisfaz $A\ge16/13$), obtém-se

   $
     T(n)\le2\,n^{2}.
   $

Para o limite inferior, nota-se diretamente que

$
T(n)=3\,T(n/4)+n^{2}\ge n^{2},
$

logo $T(n)=\Omega(n^2)$.

**Conclusão**

Como $T(n)=O(n^2)$ e $T(n)=\Omega(n^2)$, temos:

$
\boxed{T(n)=\Theta(n^2).}
$

## Q5

Recorrências dos 3 algoritmos:

* A resolve em 8 subproblemas de tamanho $n/4$ e combinação $O(1)$:
  $T_A(n)=8\,T_A(n/4)+O(1)$.

* B resolve em 4 subproblemas de tamanho $n/2$ e combinação $\Theta(n)$:
  $T_B(n)=4\,T_B(n/2)+\Theta(n)$.

* C resolve em 2 subproblemas de tamanho $n-1$ e combinação $\Theta(n^2)$:
  $T_C(n)=2\,T_C(n-1)+\Theta(n^2)$.



**Algoritmo A**

Aplicando o Mestre com $a=8$, $b=4$ e $f(n)=O(1)$:

1. Temos $n^{\log_ba}=n^{\log_4 8}=n^{1.5}$.
2. Considerando que $f(n)=O(1)=O(n^{1.5-\varepsilon})$ para qualquer $\varepsilon<1.5$.
3. Cai no Caso 1 do Mestre, logo:

   $
     T_A(n)=\Theta\bigl(n^{1.5}\bigr).
   $



**Algoritmo B**

Agora com $a=4$, $b=2$ e $f(n)=n$:

1. $n^{\log_ba}=n^{\log_2 4}=n^2$.
2. Como $f(n)=n=O(n^{2-\varepsilon})$ para $\varepsilon=1$, é Caso 1 do Mestre.
3. Conclui-se:

   $
     T_B(n)=\Theta\bigl(n^2\bigr).
   $



**Algoritmo C**

Aqui o custo satisfaz

$
T_C(n)=2\,T_C(n-1)+n^2.
$

Desenrolando a recursão:

$
\begin{aligned}
T_C(n)&=2T_C(n-1)+n^2,\\
     &=2\bigl(2T_C(n-2)+(n-1)^2\bigr)+n^2
      =2^2T_C(n-2)+2(n-1)^2+n^2,\\
     &\;\dots
      =2^{n-1}T_C(1)+\sum_{k=2}^{n}2^{\,n-k}\,k^2.
\end{aligned}
$

O termo $\sum 2^{n-k}k^2$ cresce como $2^n$, portanto

$
T_C(n)=\Theta\bigl(2^n\bigr).
$



**Conclusão**

Comparando os três crescimentos:

* $T_A(n)=\Theta(n^{1.5})$
* $T_B(n)=\Theta(n^2)$
* $T_C(n)=\Theta(2^n)$


$ \Theta(2^n) > \Theta(n^2) > \Theta(n^{1.5}) $


O **Algoritmo A** é o mais eficiente no pior caso, pois cresce mais devagar que $n^2$ e muito mais devagar que $2^n$.

**2. Função `majority_candidate(cartões)` — Divide e Conquista**

```
Função majority_candidate(cartões: lista de texto) -> texto ou nulo:

    n = tamanho(cartões)                                         # Complexidade: constante

    Se n == 1:                                                # Complexidade: constante

        retornar cartões[0]                                    # Complexidade: constante



    meio = n dividido_por_inteiro 2                              # Complexidade: constante

    esquerda = majority_candidate(cartões[0:meio])                # Complexidade: T(n/2)

    direita = majority_candidate(cartões[meio:n])                 # Complexidade: T(n/2)



    Se esquerda != nulo e direita != nulo e equivalent(esquerda, direita):   # Complexidade: 3 * T_test + constante

        retornar esquerda                                       # Complexidade: constante

    Se esquerda != nulo e direita == nulo:                    # Complexidade: constante

        retornar esquerda                                       # Complexidade: constante

    Se direita != nulo e esquerda == nulo:                    # Complexidade: constante

        retornar direita                                       # Complexidade: constante



    Para cada candidato em (esquerda, direita):                 # Loop com no máximo 2 iterações

        Se candidato == nulo:                                  # Complexidade: constante

            próximo                                           # Complexidade: constante

        contador = 0                                          # Complexidade: constante

        Para cada x em cartões:                                # Loop com 'n' iterações

            Se equivalent(candidato, x):                      # Complexidade: T_test

                contador = contador + 1                       # Complexidade: constante

        Se contador > n dividido_por_inteiro 2:                # Complexidade: constante

            retornar candidato                                  # Complexidade: constante



    retornar nulo                                             # Complexidade: constante

```

**3. Função principal `main()`:**
```
Função principal():

    cartões = argumentos_da_linha_de_comando[1:fim]           # Complexidade: O(número de argumentos) = O(n)

    Se não cartões:                                            # Complexidade: constante

        imprimir 0                                            # Complexidade: constante

        retornar                                              # Complexidade: constante



    candidato = majority_candidate(cartões)                    # Complexidade: a determinar

    Se candidato == nulo:                                      # Complexidade: constante

        imprimir 0                                            # Complexidade: constante

        retornar                                              # Complexidade: constante



    contador = 0                                              # Complexidade: constante

    Para cada x em cartões:                                # Loop com 'n' iterações

        Se equivalent(candidato, x):                          # Complexidade: T_test

            contador = contador + 1                           # Complexidade: constante



    Se contador > tamanho(cartões) dividido_por_inteiro 2:     # Complexidade: constante

        imprimir 1                                            # Complexidade: constante

    Senão:

        imprimir 0                                            # Complexidade: constante


```

**4. Calculo da Complexidade:**


`equivalent()`: $O(T)$  &nbsp;&nbsp;&nbsp;&nbsp;(T = complexidade externa)

`majority_candidate()`:

Recorrência: $T(n) = 2T(n/2) + O(n \cdot T)$

- Melhor caso ($T = O(1)$): $O(n \log n)$  
- Pior caso ($T > O(1)$): $O(n \cdot T \cdot \log n)$

`main()`:

- Melhor caso ($T = O(1)$): $O(n) + O(n \log n) = O(n \log n)$  
- Pior caso ($T > O(1)$): $O(n \cdot T) + O(n \cdot T \cdot \log n) = O(n \cdot T \cdot \log n)$

**Complexidade final:**

- Melhor caso: $O(n \log n)$  
- Pior caso: $O(n \cdot T \cdot \log n) \Rightarrow$ considerando a função externa como constante, temos $O(n \log n)$
